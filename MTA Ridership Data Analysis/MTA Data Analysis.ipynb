# Date is in object format so we will need to convert it in date format


```python
data['Date'] = pd.to_datetime(data['Date'], errors ='coerce')
data['Date'].info() #verify the conversion
```

# identify pre and post pandemic data 

```python
pandemic_start = pd.Timestamp('2020-03-01')
pandemic_stop = pd.Timestamp('2023-05-05')
pre_pandemic = data[(data['Date']>=pandemic_start) & (data['Date'] < pandemic_stop)]
post_pandemic = data[data['Date']>=pandemic_stop]
post_pandemic
```

```python
post_pandemic
```

```python
avg_pre_Pandemic = pre_pandemic.mean(numeric_only = True)
avg_pre_Pandemic
```

```python
avg_post_Pandemic = post_pandemic.mean(numeric_only = True)
avg_post_Pandemic
```

# Change in the ridership 

```python
change_in_ridership = avg_post_Pandemic - avg_pre_Pandemic
change_in_ridership
```

```python
percentage_Ridership = (change_in_ridership / avg_pre_Pandemic)*100
percentage_Ridership
```

```python
# Visualise the changes
import matplotlib.pyplot as plt

# Example: Plot changes in subway ridership
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Subways: Total Estimated Ridership'], label='Subway Ridership')
plt.axvline(pandemic_start, color='red', linestyle='--', label='Pandemic Start')
plt.axvline(pandemic_stop, color='green', linestyle='--', label='Pandemic Stop')
plt.xlabel('Date')
plt.ylabel('Ridership')
plt.title('Subway Ridership Before and After Pandemic')
plt.legend()
plt.show()
```

```python
# let us plot based on year 
data['Year'] = data['Date'].dt.year
data.head(2)
```

```python
#set the figure for subplots
fig,axes = plt.subplots(4,2, figsize=(14,10)) #we can adjust the grid size
fig.suptitle('Yearly Ridership Analysis')

#list of columns to plot 
columns_to_plot = ['Subways: Total Estimated Ridership','Buses: Total Estimated Ridership',
                  'LIRR: Total Estimated Ridership','Metro-North: Total Estimated Ridership',
                  'Access-A-Ride: Total Scheduled Trips','Bridges and Tunnels: Total Traffic','Staten Island Railway: Total Estimated Ridership']

#Loop through columns and create bar plots
for ax, col in zip(axes.flatten(), columns_to_plot):
    sns.barplot(data=data, x = 'Year', y=col, ax = ax, errorbar= None, hue='Year',palette='Set1',legend=False)
    ax.set_title(f'{col} by Year')
    ax.set_xlabel('Year')
    ax.set_ylabel('Ridership/Traffic')
    ax.tick_params(axis='x', rotation=45)
plt.tight_layout(rect=[0,0,1,0.96]) #adjust spacing to fit the title
plt.show()
```
