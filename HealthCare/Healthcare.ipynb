```python
for col in df.columns:
    if df[col].dtype == object:
        print(col,df[col][df[col] == "?"].count())
    
```

```python
#let us drop the row with missing values from column reace,gender,diag_1, diag_2, diag_3

# Identify rows to drop
drop_Idx = set(df[(df['diag_1'] == '?') & (df['diag_2'] == '?') & (df['diag_3'] == '?')].index)

# Union with additional rows based on other conditions
drop_Idx = drop_Idx.union(set(df[df['diag_1'] == '?'].index))
drop_Idx = drop_Idx.union(set(df[df['diag_2'] == '?'].index))
drop_Idx = drop_Idx.union(set(df[df['diag_3'] == '?'].index))
drop_Idx = drop_Idx.union(set(df[df['race'] == '?'].index))
drop_Idx = drop_Idx.union(set(df[df['discharge_disposition_id'] == 11].index))
drop_Idx = drop_Idx.union(set(df[df['gender'] == 'Unknown/Invalid'].index))

# Ensure indices are within bounds
drop_Idx = [idx for idx in drop_Idx if idx in df.index]

# Create list of valid indices
new_Idx = list(set(df.index) - set(drop_Idx))

# Check if there are any valid indices left
if not new_Idx:
    raise ValueError("No valid indices remaining after filtering.")

# Filter DataFrame to keep only valid rows
df = df.iloc[new_Idx]

# Reset index to ensure itâ€™s continuous and clean
df = df.reset_index(drop=True)

```
